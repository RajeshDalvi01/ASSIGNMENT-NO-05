# MULTIPLE LINEAR REGRESSION

Multiple linear regression (MLR), also known simply as multiple regression, is a statistical technique that uses several explanatory variables to predict the outcome of a response variable. Multiple regression is an extension of linear (OLS) regression that uses just one explanatory variable.

OUR GOAL IS TO SELECT BEST MODEL USING REQUIRED AND MINIMUM NO. OF INPUT ( INDEPENDANT ) VARIABLES TO PREDICT THE ACCURATE TARGET(DEPENDANT) VARIABLE.FOR THAT, FIRSTLY WE HAVE TO CHECK CORRELATION BETWEEN INDEPENDANT VARIABLES AND CHOOSE REQUIRED NO. OF VARIABLES FOR BUILDING THE MODEL WITH HIGHER ACCURACY.
